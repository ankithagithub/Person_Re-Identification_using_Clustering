```python
import numpy as np
from keras.preprocessing import image
from keras.applications.vgg16 import VGG16, preprocess_input
from keras.models import Model

# Load the pre-trained VGG16 model
base_model = VGG16(weights='imagenet', include_top=True)
# Remove the classification layer
model = Model(inputs=base_model.input, outputs=base_model.get_layer('fc2').output)

# Load an example image
img_path = '/mnt/DATA/Ankita/spc24/cropped_objects_uniform-new/1.jpg'  # Replace 'example_image.jpg' with the path to your image
img = image.load_img(img_path, target_size=(224, 224))
x = image.img_to_array(img)
x = np.expand_dims(x, axis=0)
x = preprocess_input(x)

# Extract features
features = model.predict(x)

# Reshape to a vector
features_vector = features.flatten()

# Print the dimensions of the feature vector
print("Feature vector dimensions:", features_vector.shape)

# Store the feature vector in x1
x1 = features_vector
```

```python
import numpy as np
from keras.preprocessing import image
from keras.applications.resnet50 import ResNet50, preprocess_input
from keras.models import Model

# Load the pre-trained ResNet50 model
base_model = ResNet50(weights='imagenet', include_top=False)
# Extract features from the penultimate layer
model = Model(inputs=base_model.input, outputs=base_model.layers[-2].output)

# Load an example image
img_path = '/mnt/DATA/Ankita/spc24/cropped_objects_uniform-new/1.jpg'  # Replace 'example_image.jpg' with the path to your image
img = image.load_img(img_path, target_size=(224, 224))
x = image.img_to_array(img)
x = np.expand_dims(x, axis=0)
x = preprocess_input(x)

# Extract features
features = model.predict(x)

# Reshape to a vector
features_vector = features.flatten()

# Print the dimensions of the feature vector
print("Feature vector dimensions:", features_vector.shape)

# Store the feature vector in x2
x2 = features_vector
```

```python
from skimage.feature import hog
from skimage import exposure
from skimage import io
from skimage.color import rgb2gray

# Load an example image
img_path = '/mnt/DATA/Ankita/spc24/cropped_objects_uniform-new/1.jpg'  # Replace 'example_image.jpg' with the path to your image
image = io.imread(img_path)

# Convert the image to grayscale if it's not already
if len(image.shape) > 2:
    image = rgb2gray(image)

# Compute HOG features
fd, hog_image = hog(image, orientations=8, pixels_per_cell=(16, 16),
                    cells_per_block=(1, 1), visualize=True)

# Reshape HOG features to a vector
features_vector = fd.flatten()

# Print the dimensions of the feature vector
print("HOG feature vector dimensions:", features_vector.shape)

# Store the feature vector in x3
x3 = features_vector
```
